```python
import pandas as pd
```

```python
df = pd.read_csv('cenlab-full.csv')
```

```python
df2 = pd.read_csv('txtlab/txtlab-meta.csv')
```

```python
def fixAuthor(author): 
    e = eval(author)
    return e[1] + ' ' + e[0]
```

```python
df2['author'] = df2['author'].apply(fixAuthor)
```

```python
unified = pd.concat([df, df2])
```

```python
unified.to_csv('cenlab-everything.csv')
```
